## Name of Project : Image to Pencil Sketch with Python

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
image = cv2.imread("xyz.jpg")
```

```python
image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(image))
```

```python
grayImg = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(grayImg))
```

```python
inv_grayImg = 255 - grayImg
```

```python
inv_grayImg
```

```python
display(Image.fromarray(inv_grayImg))
```

```python
blur_img = cv2.GaussianBlur(inv_grayImg, (35,35), 0)
```

```python
display(Image.fromarray(blur_img))
```

```python
inv_blurImg = 255 - blur_img
```

```python
inv_blurImg
```

```python
display(Image.fromarray(inv_blurImg))
```

```python
def mix(grayImg, blur_img):
    return cv2.divide(grayImg, inv_blurImg, scale = 210.0)
```

```python
pencil_sketch = mix(grayImg, inv_blurImg)
display(Image.fromarray(pencil_sketch))
```
